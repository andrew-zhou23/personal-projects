## Section 5: Data Science

The question that I choose to explore in this section is whether or not I can predict how many points a player will average, based on age, minutes played, and field goals attempted per game. I plan to create a linear regression model that will predict how many points an NBA player will average. The reason that I chose more categories than just age is because there isn't a strong correlation between the two. Sure, some players follow the script that older players should score more points, however there are many outliers, thus there isn't a strong enough correlation. Additionally, minutes played and field goals attempted contribute to points per game because the more minutes you play, the more opportunities you get to score, and the more shots you take, the more likely you are to score. Also, after creating a correlation matrix, I can see that minutes played has as 0.88 correlation, field goals attempted has a 0.98 correlation, and age has a correlation of 0.14. So, since age has a weak correlation, adding two more independent variables that have a strong correlation with points per game will allow the model to work effectively. First, I need to establish what the independent variables and the dependent variable are. Next, I need to train a model and then use the model by using model.predict.
Then, I am going to add a column to the data frame that has a prediction for the points per game a player will average and a column to the data frame that contains the residuals of my model (PTS - Predicted Points).

```python
df.corr(numeric_only=True)
```

```python
from sklearn.linear_model import LinearRegression
ind = ['Age', 'MP', 'FGA']
dep = "PTS"
train = df.sample(frac = 0.7)
model = LinearRegression()
model = model.fit(train[ind], train[dep])
df["Predicted Points"] = model.predict(df[ind])
df["Residual"] = df["PTS"] - df["Predicted Points"]
df
```

## Section 6: A Different Data Visualization

The histogram illustrates how my model was actually pretty effective in predicting the points per game of NBA players. This means that my model was effectively predicting the points per game for NBA players because the residual is calculated by subtracting the predicted points from the actual points.
